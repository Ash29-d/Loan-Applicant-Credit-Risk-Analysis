## 5) Split the data in training and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(a,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
tree=DecisionTreeClassifier(criterion='gini',max_depth=6,random_state=30, splitter='best')
```

```python
tree.fit(x_train,y_train)
```

```python
tree.score(x_test,y_test)
```

```python
tree.score(x_train,y_train)
```

```python
plt.figure(figsize=[20,20])
plot_tree(tree,filled=True,feature_names=x.columns);
```
